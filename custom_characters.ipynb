```python
import os
import asyncio
from convex import ConvexClient

async def get_memories():
    """
    Retrieves all memories from the Convex database.
    """
    # Get the Convex deployment URL from an environment variable.
    # You can find this in your Convex dashboard or by running `npx convex url`.
    url = os.environ.get("CONVEX_URL")
    if not url:
        raise ValueError("CONVEX_URL environment variable not set.")

    print(f"Connecting to Convex at {url}...")
    client = ConvexClient(url)

    try:
        # Fetch all documents from the "memories" table.
        print("Fetching memories...")
        memories = await client.query("memories:listMemories")
        print(f"Found {len(memories)} memories.")
        for i, memory in enumerate(memories):
            print(f"\n--- Memory {i+1} ---")
            print(f"  ID: {memory['_id']}")
            print(f"  Player ID: {memory['playerId']}")
            print(f"  Description: {memory['description']}")
            print(f"  Importance: {memory['importance']}")
            print(f"  Last Accessed: {memory['lastAccess']}")
            # You can add more fields here if you need them.

    except Exception as e:
        print(f"An error occurred: {e}")
    finally:
        # It's important to close the client when you're done.
        client.close()

asyncio.run(get_memories())
```
